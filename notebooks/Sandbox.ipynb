# <span style="color: red;"> High-level MAE for Taxi 5x5 room, lambda 1</span>

```python
res_v1 = get_MAE_HL_errors(f'{results_dir}/taxi_5_v1_333.pkl', '$V_1$-IL', 5000, 10, (8, 5))
res_v2 = get_MAE_HL_errors(f'{results_dir}/taxi_5_v2_333.pkl', '$V_2$-IL', 5000, 10, (8, 5))
res_v3 = get_MAE_HL_errors(f'{results_dir}/taxi_5_v3_333.pkl', '$V_3$-IL', 5000, 10, (8, 5))
res_plain = get_MAE_plain_Z_learning(f'{results_dir}/taxi_5_flat_333.pkl', 'Z-learning+IS', 5000, 10)

filename = 'overall_taxi_5_1.pdf'

csfont = {'fontname':'Helvetica'}
hfont = {'fontname':'Helvetica', 'fontsize':12}

plt.close('all')

fig, ax = plt.subplots(1,1, figsize=(9, 6))

lambda_ = 1

ax.set_title(fr'Taxi domain. $ 5\times5 $ room, $\lambda={lambda_}$', **csfont)
       
ax.plot(res_v1[2], linewidth=1, markevery=250, color='gray', marker='x')
ax.plot(res_v2[2], linewidth=1, markevery=250, color='brown', marker='x')
ax.plot(res_v3[2], linewidth=1, markevery=250, color='royalblue', marker='^')
ax.plot(res_plain, linewidth=1, markevery=250, color='black', marker='o')

plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

legend = ['$V_1$', '$V_2$', '$V_3$', 'Z-IS']

ax.legend(legend)
ax.set_xlim((0,5000))
ax.set_ylim((0,20))

ax.set_ylabel('MAE in $\hat V(s)$', **hfont)
ax.set_xlabel('samples', **hfont)
plt.savefig(filename, bbox_inches='tight', dpi=500)
plt.show()
```
